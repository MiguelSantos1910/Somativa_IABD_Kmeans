```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_samples, silhouette_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/danielvieira95/Intelig-ncia-Artificial-e-Big-Data--2-2024/refs/heads/master/Bases%20de%20dados/dadosacoes.csv')
```

```python
dados.describe()
```

```python
dados_novos = pd.get_dummies(dados, columns=['nome ação'])
```

```python
dados_novos
```

```python
#Boxplot - Preço Ação
sns.boxplot(x = dados['nome ação'], y = dados['preço ação R$'])
```

```python
#Boxplot - Valor de Mercado
sns.boxplot(x = dados['nome ação'], y = dados['valor de mercado R$ -(Bilhões)'], color='r')
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(dados_novos)
```

```python
sse = kmeans.inertia_
print(f"SSE: {sse}")
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dados['nome ação'], dados['preço ação R$'], c=labels)
plt.scatter(centroids[:, 4], centroids[:, 4], marker='x', color='green')
plt.xlabel('Nome Ação')
plt.ylabel('Preço Ação (R$)')
print('Gráfico com 4 clusters')
```

```python
dados_novos['Grupo'] = labels
grupo_data = dados_novos.groupby('Grupo').mean()
print(f'Grupo: {grupo_data}')
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dados['preço ação R$'], dados['qtde cotas'], dados['valor de mercado R$ -(Bilhões)'], c=labels)

ax.set_title('KMeans clusters encontrados')
ax.set_xlabel('Preço Ação (R$)')
ax.set_ylabel('Quantidade de Cotas')
ax.set_zlabel('Valor de Mercado R$ - (Bilhões)')
plt.show()

print('Gráfico com 4 clusters')
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(dados_novos)
graficos = []

for k in range(1, 9):
    km = KMeans(n_clusters=k, init="k-means++", random_state=42, n_init='auto')
    km.fit(X_scaled)
    graficos.append(km.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 9), graficos, marker='o')
plt.title("Método do cotovelo — WCSS x número de clusters")
plt.xlabel("Número de clusters k")
plt.ylabel("WCSS (inércia)")
plt.xticks(range(1, 9))
plt.grid(True)
plt.show()
```

```python
scores = []

for k in range(1, 9):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    scores.append(score)
    print(f"k={k} → silhueta média = {score}")

plt.figure(figsize=(8,5))
plt.plot(range(2, 9), scores, marker='o')
plt.xticks(range(2, 9))
plt.title("Silhueta média por número de clusters")
plt.xlabel("Número de clusters k")
plt.ylabel("Silhueta média")
plt.grid(True)
plt.show()
```
